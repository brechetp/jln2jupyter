```julia
## Generate statistics on matrices with Gaussian entries.# Eigenvalues of matrices where each entry is a random number from the sample# distribution over have interesting properties in their eigenvalue spectra.# Here, we will look at this set with 500 by 500 matrices and 100 samples,# where each entry is a normal random variable.n = 500 # generate 500x500 random matrixt = 100 # use 100 random samplessample(n) = eigvals!(randn(n,n)) # define the sample functionsamples = map(_ -> sample(n), 1:t)evals = collect(chain(samples...))
```

```julia
## Plot the resultshistogram2d(evals, nbins=50)xlabel!("Re(λ)")ylabel!("Im(λ)")title!("Eigenvalue Histogram")
```
